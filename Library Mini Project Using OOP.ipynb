```python
class Library:
    
    def __init__(self, booklist, lib_name, user, book):
        self.booklist = booklist
        self.lib_name = lib_name
        self.user = user
        self.book = book
        self.book_dict = dict()
        
    def DisplayBook(self):
        for book in self.booklist:
            print(book)
    
    def LendBook(self):
        if self.book in self.booklist:
            if self.book not in self.book_dict:
                self.book_dict.update({self.book: self.user})
                self.booklist.remove(self.book)
                print("The Library Has Been Updated!! Take The Book")
            else:
                print("The book is not available in the library")
        else:
            print("The book is not available in the library")
            
    def AddBook(self, book_new):
        self.booklist.append(book_new)
        
    def ReturnBook(self):
        self.book_dict.pop(self.book)
        self.booklist.append(book)
        print(f"The book: {self.book} has been returned by the user: {self.user}")
        
        
            
```

```python
l1 = Library(['Harry Potter', 'Eragon', 'Lord Of The Rings', 'Alex Rider'], "RK's Library", "name", "book")
print(l1.lib_name)
print()
print("1. Display The Books")
print("2. Lend The Books")
print("3. Add The Book")
print("4. Return The Book")
print("\n"*2)
var = True
while var:
    
    user_choice1 = int(input("Enter Your Choice [1, 4]: "))
    print()
    if user_choice1 == 1:
        l1.DisplayBook()
    elif user_choice1 == 2:
        name = input("Enter The Name: ")
        book = input("Enter The Book Name: ")
        print()
        l1.user = name
        l1.book = book
        l1.LendBook()
    elif user_choice1 == 3:
        print()
        book_new = input("Enter A New Book: ")
        l1.AddBook(book_new)
    elif user_choice1 == 4:
        l1.ReturnBook()
    else:
        print()
        print("Wrong Choice")
        print("\n"*2)
    
    user_choice2 = ""
    while user_choice2 != 'y' and user_choice2 != 'n':
        print()
        user_choice2 = input("Do You Want To Continue (y Or n): ").lower().strip()
        if user_choice2 == 'y':
            var = True
        elif user_choice2 == 'n':
            var = False
            break
        else:
            print()
            print("Wrong Choice"+"\n")
        
```
