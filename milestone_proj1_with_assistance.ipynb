## Tic Tac Toe



```python
# To Display Board disp_boar(mylist):

from IPython.display import clear_output
def disp_board(mylist):
    clear_output()
    print(mylist[7] + "|" + mylist[8] +"|" + mylist[9])
    print("-+-+--")
    print(mylist[4] + "|" + mylist[5] +"|" + mylist[6])
    print("-+-+--")
    print(mylist[1] + "|" + mylist[2] +"|" + mylist[3])
    
mylist = [" "]*10

```

```python
# To take 'X' Or 'O' from User

def player_input():
    marker = " "
    while marker != 'X' and marker != 'O':
        marker = input("Player Enter X or O: ").upper()
    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
```

```python
#To Check for Win win(mylist, marker):

def win(mylist, marker):
    return ((mylist[1] == mylist[2] == mylist[3] == marker) or 
            (mylist[4] == mylist[5] == mylist[6] == marker) or 
            (mylist[7] == mylist[8] == mylist[9] == marker) or 
            (mylist[1] == mylist[4] == mylist[7] == marker) or 
            (mylist[2] == mylist[5] == mylist[8] == marker) or
            (mylist[3] == mylist[6] == mylist[9] == marker) or
            (mylist[1] == mylist[5] == mylist[9] == marker) or
            (mylist[3] == mylist[5] == mylist[7] == marker))
```

```python
#To check for spaces 

def space_check(mylist, pos):
    return mylist[pos] == " "
```

```python
# To check for full board

def check_full_board(mylist):
    for i in range(1, 10):
        if space_check(mylist, i):
            return False
    
    return True
    
```

```python
# To check for player choice of position

def player_choice(mylist):
    position = 0
    while position not in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
        position = int(input("Player Choose A Position From (1 - 9): "))
        if space_check(mylist, position):
            return position
        else:
            print("Position already occupied choose again")
            player_choice(mylist)
        
```

```python
# To see who goes first

from random import *

def toss():
    winner = randint(0, 1)
    if winner == 1:
        return "Player 1"
    else:
        return "Player 2"
```

```python
# To see if the player/s want to play again

def replay():
    choice = input("Do You Want To Play Again 'yes' Or 'no': ").lower()
    return choice == 'yes'
```

```python
# In case of single player. Computer Moves

from random import *
def comp_pos(mylist):
    position = randint(1, 9)
    #print(position)
    if space_check(mylist, position) == True:
        return position
    else:
        comp_pos(mylist)
```

```python
# To Place Marker On The Board Takes Three Arguments

def place_marker(mylist, marker, pos = 0):
    mylist[pos] = marker
#place_marker(test_board, "$", 6)
#disp_board(test_board)
```
